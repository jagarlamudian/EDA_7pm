**Fill with methods**

- whenever we are using methods no need provide values.

```python
import pandas as pd
import numpy as np
Data={'Names':['Cool','Silent','Decent',np.nan],
      'Age':[np.nan,22,25,27],
      'City':['hyd','gnt',np.nan,'nrt']}


df=pd.DataFrame(Data)
df
```

- pad
- bfill
- ffill
- backfill

```python
import warnings
warnings.filterwarnings('ignore')
df.fillna(method='pad')
```

```python
import warnings
warnings.filterwarnings('ignore')
df.fillna(method='bfill')
```

```python
import warnings
warnings.filterwarnings('ignore')
df.fillna(method='ffill')
```

```python
import warnings
warnings.filterwarnings('ignore')
df.fillna(method='backfill')
```

```python
age_median=df['Age'].median()
df['Age'].fillna(age_median)
```

```python
age_mean=df['Age'].mean()
df['Age'].fillna(age_mean)
```

```python
age_mode=df['Age'].mode()
df['Age'].fillna(age_mode)
```

**KNN Immputer**

- K-Nearest Neighbours
- where k = hyper parameter which means the user can change
- idea:Instead of taking all the samples average
    - First fix the neighbours number i.e.k
    - Then find the neighbours using distance metric
    - Then take the average of those samples to fill the missing values

```python
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import KNNImputer
```
